```python
age = int(input("Enter your age : "))
weight = int(input("Enter your weight : "))
height = input("Enter your height(in cm)  : ")
activity = input("Enter your activity level (low, moderate, high) : ")
diet = input("Enter your diet performance (vegetarian, non-vegetarian) : ")
fitness_goal = input("Enter your goal (fat loss , fat gain) : ")
```

```python
user_profile = {
  "age": age,
  "weight_kg": weight,
  "height_cm": height,
  "activity_level": activity,
  "diet_preference": diet,
  "fitness_goal": fitness_goal
}

user_query = """Collect user details (age, weight, height, activity level, diet preference, fitness goal)

Use a LangChain multi-step workflow to generate:

üçΩÔ∏è Meal plan with macros (protein/carbs/fats)
üèÉ Workout plan with warm-up, main workout, cool-down
üìÖ Weekly combined strategy with schedule + motivation tips
Display results neatly inside Colab (Markdown-style output is enough)"""
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from google.colab import userdata
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    google_api_key = userdata.get("GEMINI_API_KEY")
)

print(llm.invoke("Hi,who are you ?"))
```

**Meal chain**

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

meal_prompt = PromptTemplate(
    template="Based on the following user profile, please generate a detailed meal plan with macros (protein/carbs/fats) for a day:\n\n{user_profile}\n\nMeal Plan:",
    input_variables=["user_profile"]
)

meal_chain = meal_prompt | llm | StrOutputParser()

result = meal_chain.invoke({"user_profile": user_profile})

print(result)
```

**Workout Chain**

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

workout_prompt = PromptTemplate(
    template="Based on the following user profile, please generate a detailed workout plan with warm-up, main workout, cool-down for a day:{user_profile}Meal Plan:",
    input_variables=["user_profile"]
)

workout_chain = workout_prompt | llm | StrOutputParser()

result = workout_chain.invoke({"user_profile": user_profile})

print(result)
```
