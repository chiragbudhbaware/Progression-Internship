**✅ Problem 1: Collect Student Details (Variables + Input + List)**

  

```python
name = input("Enter your name : ")
physics = int(input("Enter the physics marks : "))
chemistry = int(input("Enter the chemistry marks : "))
maths = int(input("Enter the maths marks : "))

marks = [physics,chemistry,maths]

print(f"Your name is : {name}")
print(marks)
```

 **✅Problem 2: Calculate Total and Average (Loop + List)**



```python
marks = [physics,chemistry,maths]
total_marks = 0

for mark in marks:
  total_marks += mark
print(f"Your total marks is : {total_marks}")

average_marks = total_marks/len(marks)
print(f"Your average marks is : {average_marks}")
```

 **✅ Problem 3: Decide Pass or Fail (Conditionals + Loop)**

```python
marks = [physics,chemistry,maths]
count = 0
result = "Fail"

for mark in marks:
  if mark < 35:
    result
    break

  else:
    count += 1
    if count == len(marks):
      result = "Pass"

print(result)


```

**✅ Problem 4: Create a Student Dictionary (Dictionary + Key-Value Storage)**

```python
student = {
    "name":name,
    "marks":marks,
    "total":total_marks,
    "average":average_marks,
    "result":result
}

print(student)
```
